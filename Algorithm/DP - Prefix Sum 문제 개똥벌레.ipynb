* **종유석:** 천장에서부터 자라는 장애물
* (무조건 석순 → 종유석 → 석순 → 종유석… 순서로 등장)

## 예시 풀이

```python
import sys
import io

def run_my_solution(input_data):
    # 코랩 환경에서 백준과 똑같이 sys.stdin과 print를 쓰기 위한 트릭
    original_stdin = sys.stdin
    original_stdout = sys.stdout
    sys.stdin = io.StringIO(input_data)
    sys.stdout = io.StringIO()

    try:
        # ========================================================
        # [여기에 백준 제출용 코드를 그대로 작성하세요]

        n, h = map(int, sys.stdin.readline().split())

        # 누적합 맨 끝에 0을 패딩할 것이기에 높이 + 1 로 받음
        s = [0] * (h + 1)  # 석순
        j = [0] * (h + 1)  # 종유석

        min_count = n  # 파괴해야 하는 장애물의 최소값
        range_count = 0  # 최소값이 나타나는 구간의 수

        # 1. 장애물의 높이에 해당하는 칸에 개수를 누적
        for i in range(n):
            if i % 2 == 0: # 짝수는 석순
                s[int(input())] += 1
            else: # 홀수는 종유석
                j[int(input())] += 1

        # 2. 역방향 누적 합 (h2 > h1 이고, h2에 장애물이 있다면 h1에는 장애물이 있던 없든 부딪히니까)
        for i in range(h - 1, 0, -1):
            s[i] += s[i + 1]
            j[i] += j[i + 1]

        # 3. 최소 충돌 횟수 및 구간 구하기
        for i in range(1, h + 1):
            # i 높이에서 부딪히는 총 장애물 수
            # 종유석은 거꾸로 계산하는거 주의
            current_count = s[i] + j[h - i + 1]

            if min_count > current_count:
                min_count = current_count
                range_count = 1
            elif min_count == current_count:
                range_count += 1

        print(min_count, range_count)

        # ========================================================
    except Exception as e:
        print(f"Error: {e}")
    finally:
        # 출력 결과 수집 및 원래 환경으로 복구
        output_data = sys.stdout.getvalue().strip()
        sys.stdin = original_stdin
        sys.stdout = original_stdout

    return output_data
```
